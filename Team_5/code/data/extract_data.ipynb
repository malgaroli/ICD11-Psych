```python
import os
import re
import pandas as pd
from docx import Document
```

```python
# Define paths to the folders containing the .docx files
folders = {
    'Anxiety': 'anxiety/vignettes_Anxiety',
    'Mood': 'Mood/vignettes_Mood',
    'Stress': 'Stress/vignettes_Stress'
}
output_data = []

def extract_vignette_data(doc_path, category):
    """Extract relevant sections from a docx file and add category."""
    # Extract Vignette ID from filename
    vignette_id = os.path.splitext(os.path.basename(doc_path))[0]
    
    # Open the .docx file
    doc = Document(doc_path)
    text = [para.text for para in doc.paragraphs if para.text.strip() != '']

    # Define pattern markers for the sections
    data = {
        'Category': category,
        'Vignette ID': vignette_id,
        'Name': '',
        'Referral': '',
        'Presenting Symptoms': '',
        'Additional Background Information': ''
    }
    section = None

    for para in text:
        # Detect sections with or without colon
        if re.match(r"^Referral[:]*$", para, re.IGNORECASE):
            section = 'Referral'
        elif re.match(r"^Presenting Symptoms[:]*$", para, re.IGNORECASE):
            section = 'Presenting Symptoms'
        elif re.match(r"^Additional Background Information[:]*$", para, re.IGNORECASE):
            section = 'Additional Background Information'
        elif re.match(r"^[A-Z][A-Za-z]*$", para):  # To capture the Name section, assuming it's a single capitalized word.
            data['Name'] = para
        elif section:  # Append paragraph text to the current section
            data[section] += " " + para.strip()
    
    return data
```

```python
# Iterate over the defined folders and process files within each category
for category, folder_path in folders.items():
    for filename in os.listdir(folder_path):
        if filename.endswith(".docx"):
            file_path = os.path.join(folder_path, filename)
            vignette_data = extract_vignette_data(file_path, category)
            output_data.append(vignette_data)

# Convert to DataFrame and save to CSV
df = pd.DataFrame(output_data)
output_file = 'Data.csv'
df.to_csv(output_file, index=False)

print(f"Extraction complete. Data saved to {output_file}")
```
